## Download and Save Offline Directories

### Homepage of Castbox.fm

```python
# Save offline copy of homepage. Scraped Sept 30, 2019
cb_home_html = requests.get('https://castbox.fm/home?country=us').text
with open("../scraped/category/castbox_home.html", "w") as file:
    file.write(cb_home_html)
cb_home_soup = BeautifulSoup(cb_home_html, 'lxml')
```

```python
# Test that channel file is saved and read properly
with open("../scraped/category/castbox_home.html", "r") as file:
    cat_soup = BeautifulSoup(file, 'lxml')

print(cat_soup.title)
print(cat_soup.find('h1'))
print(cat_soup.find('a'))
```

```python
# Aggregate all podcasts from the main page.
# Collect Name, Channel Link, Publisher, Cover Image Link


def collect_home_pods(soup,index_dir):
    # prevent trailing slashes from malforming URL
    if index_dir[-1] == '/':
        index_dir = index_dir[:-1]
        
    # each podcast listed on the homepage will be a dictionary k,v pair
    pods_dict = {}
    
    top_categories = soup.find_all(class_='topPodcast')
    #     print(top_categories)
    
    for cat in top_categories:
        category = cat.find(class_='coverBoxTitle').text
        pod_rows = cat.find_all('div', class_='coverRow')
        for row in pod_rows:
            chan_url = index_dir + row.find('a').get('href')
    #         print(chan_url)
            title = row.find(class_='title').text
    #         print(title)
            author = row.find(class_='author').text
    #         print(author)
    #         print('\n')
            pods_dict[title] = {'chan_url': chan_url, 'author': author, 'category':category}
        
    return pods_dict
    
cb_home_dir = collect_home_pods(cat_soup, 'https://castbox.fm/')
[i for i in cb_home_dir.items()][:5]
```

```python
# dump homepage directory to json
with open('../scraped/category/cb_home_dir.json', 'w') as file:
    json.dump(cb_home_dir, file)
```

```python
# Read it back
with open('../scraped/category/cb_home_dir.json', 'r') as file:
    cb_home_dir = json.load(file)

import random

test_k = random.sample(cb_home_dir.keys(), 1)
print(test_k[0], cb_home_dir[test_k[0]])
```

### "Top Shows" (category 0) - HTML Capture

https://castbox.fm/categories/0?country=us

```python
# Save offline copy of top shows page. Scraped Sept 30, 2019
cb_topshows_html = requests.get('https://castbox.fm/categories/0?country=us').text
with open("../scraped/category/cb_topshows.html", "w") as file:
    file.write(cb_topshows_html)
cb_top_soup = BeautifulSoup(cb_topshows_html, 'lxml')
```

```python
# Test that channel file is saved and read properly
with open("../scraped/category/cb_topshows.html", "r") as file:
    cat_soup = BeautifulSoup(file, 'lxml')

print(cat_soup.title)
print(cat_soup.find('h1'))
print(cat_soup.find('a'))
```

```python
# Aggregate all podcasts from the top show page.
# Collect Name, Channel Link, Publisher, Cover Image Link


def collect_category_pods(soup, index_dir):
    '''
    Create navigational dict of podcast channel as lists on a category page,
    for the top podcasts across all categories, or a single category.
    '''
    # prevent trailing slashes from malforming URL
    if index_dir[-1] == '/':
        index_dir = index_dir[:-1]
        
    pods_dict = {}
    pod_rows = soup.find_all('div', class_='coverRow')
    
    for row in pod_rows:
        chan_url = index_dir + row.find('a').get('href')
#         print(chan_url)
        title = row.find(class_='title').text
#         print(title)
        author = row.find(class_='author').text
#         print(author)
#         print('\n')
        pods_dict[title] = {'chan_url': chan_url, 'author': author}
    
    return pods_dict
    
cb_top_dir = collect_category_pods(cat_soup, 'https://castbox.fm/')
[i for i in cb_top_dir.items()][:5]
```

```python
# dump homepage directory to json
with open('../scraped/category/cb_top_dir.json', 'w') as file:
    json.dump(cb_top_dir, file)
```

```python
# Read it back
with open('../scraped/category/cb_home_dir.json', 'r') as file:
    cb_home_dir = json.load(file)

import random

test_k = random.sample(cb_top_dir.keys(), 1)
print(test_k[0], cb_top_dir[test_k[0]])
```
